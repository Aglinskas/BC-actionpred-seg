```python
import pandas as pd
import os 
import numpy as np
import random
```

```python
ofdir = './Data/01_gifs_for_mTurk_orig/'
```

```python
infn = os.path.join(ofdir,'data_legend.csv')
```

```python
df = pd.read_csv(os.path.join(ofdir,'data_legend.csv'))
n = df.shape[0]
df
```

```python
categories = df['action'].values
```

```python
answer_choices = []
for c in range(n):
    #answer_choices.append( [random.choice(list(categories[np.arange(n)!=c])) for _ in range(5)] )
    #answer_choices.append( [random.choice(list(categories[np.arange(n)!=c])) for _ in range(5)] )
    answer_choices.append( random.sample(list(categories[np.arange(n)!=c]),5) )
answer_choices = np.array(answer_choices)
```

```python
for c in range(n):
    answer_choices[c,np.random.randint(0,5)] = categories[c]
```

```python
df['choice_1'] = list(answer_choices[:,0])
df['choice_2'] = list(answer_choices[:,1])
df['choice_3'] = list(answer_choices[:,2])
df['choice_4'] = list(answer_choices[:,3])
df['choice_5'] = list(answer_choices[:,4])
```

```python
all_5_unique = np.array([len(np.unique(df[['choice_1', 'choice_2','choice_3', 'choice_4', 'choice_5']].iloc[i].values))==5 for i in range(df.shape[0])])
```

```python
contains_correct = np.array([categories[i] in df[['choice_1', 'choice_2','choice_3', 'choice_4', 'choice_5']].iloc[i].values for i in range(df.shape[0])])
```

```python
assert all(all_5_unique),'not unique'
assert all(contains_correct),'correct mssing'
```
