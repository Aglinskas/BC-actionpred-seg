```python
import os
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
from tqdm import tqdm
```

```python
train_dir = './Data/pytorch-data/action_data_orig/train/'
test_dir = './Data/pytorch-data/action_data_seg/test/'
categories = os.listdir(train_dir)
categories.sort()
categories[0:3]
```

```python
mean_frames_test = []
for c in tqdm( range(len(categories)) ):
    cat_dir = os.path.join(test_dir,categories[c])
    frames = [frame for frame in os.listdir(cat_dir) if not frame.startswith('.')]
    frames.sort()
    cat_frames = np.array([np.array(Image.open(os.path.join(cat_dir,frame))) for frame in frames])
    mean_frames_test.append( cat_frames.mean(axis=0).astype(np.uint8) )
```

```python
mean_frames = []
for c in tqdm( range(len(categories)) ):
    cat_dir = os.path.join(train_dir,categories[c])
    frames = [frame for frame in os.listdir(cat_dir) if not frame.startswith('.')]
    frames.sort()
    cat_frames = np.array([np.array(Image.open(os.path.join(cat_dir,frame))) for frame in frames])
    mean_frames.append( cat_frames.mean(axis=0).astype(np.uint8) )
```

```python
len(mean_frames)
```

```python
plt.figure(figsize=(20,35))
for i in range(45):
    plt.subplot(9,5,i+1)
    plt.imshow(mean_frames[i])
    #plt.imshow(mean_frames_test[i])
    plt.axis('off')
    plt.title(categories[i] + ' ' + str(i),fontsize=16)
    
#plt.tight_layout()
```

```python
plt.figure(figsize=(25,5))
plt.plot([mean_frame.mean() for mean_frame in mean_frames],'.',markersize=20)
plt.xticks(np.arange(45));
```

```python
plt.figure(figsize=(25,5))
plt.plot([np.corrcoef(mean_frames[i].flatten(),np.array(mean_frames_test).mean(axis=0).flatten())[0,1] for i in range(45)],'.',markersize=20)
plt.xticks(np.arange(45));
```
